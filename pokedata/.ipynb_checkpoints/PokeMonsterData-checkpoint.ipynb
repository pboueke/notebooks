Dataset: [https://www.kaggle.com/abcsds/pokemon/](https://www.kaggle.com/abcsds/pokemon/)

Idea: get the feeling of the strengths and weakness of each type of pokemon (considering only the primary type of each pokemom).



Setup:

```python
%matplotlib inline
import os
import math
import random
import pdb
import matplotlib.pyplot as plt

class Pokebicho:
    pass

class PokeTipo:
    pass

maxid = 1
poketipos = {}
poketype_relations = {}
poketipos_stats = {}
```

Reading the files and storing the data in memory:

```python
print("Starting type loading")
with open("types.csv") as rd_input:
        tps = rd_input.readline().split(",")
        index = 0
        poketype_relations["order"] = {}
        # sets the indexation order
        for name in tps:
            poketype_relations["order"][name.strip('\n')] = index
            index += 1
        # for each type, it's attack modifier against another class can be accessed using the "order" key
        for line in rd_input:
            lnl = line.split(",")
            poketype_relations[lnl[0]] = []
            for i in range(1, len(lnl)):
                poketype_relations[lnl[0]].append(float(lnl[i]))
        
print("Starting monster loading")
with open("Pokemon.csv") as rd_input:
        rd_input.readline()
        for line in rd_input:
            lnl = line.split(",")
            pk = Pokebicho()
            pk.id = int(lnl[0])
            pk.name = lnl[1]
            pk.type1 = lnl[2]
            pk.type2 = lnl[3] if lnl[3] != "" else lnl[2]
            pk.total = int(lnl[4])
            pk.hp = int(lnl[5])
            pk.attack = int(lnl[6])
            pk.defense = int(lnl[7])
            pk.sp_attack = int(lnl[8])
            pk.sp_defense = int(lnl[9])
            pk.speed = int(lnl[10])
            if pk.id > maxid:
                maxid = pk.id
            if pk.type1 in poketipos:
                poketipos[pk.type1].append(pk)
            else:
                poketipos[pk.type1] = []
                poketipos[pk.type1].append(pk)

print("Data Loaded")
```

Let's calculate the basic statistics for each type

```python
for tipo in poketipos.keys():
    counter = 0
    pkt = PokeTipo()
    pkt.name = tipo
    pkt.names = ""
    pkt.mean_attack = 0
    pkt.mean_sattack = 0
    pkt.mean_defense = 0
    pkt.mean_sdefense = 0
    pkt.mean_hp = 0
    pkt.mean_total = 0
    pkt.mean_speed = 0
    pkt.sd_attack = 0
    pkt.sd_sattack = 0
    pkt.sd_defense = 0
    pkt.sd_sdefense = 0
    pkt.sd_hp = 0
    pkt.sd_speed = 0
    pkt.sd_total = 0
    # means
    for bixo in poketipos[tipo]:
        counter += 1
        pkt.names += bixo.name + " "
        pkt.mean_attack += bixo.attack
        pkt.mean_sattack += bixo.sp_attack
        pkt.mean_defense += bixo.defense
        pkt.mean_sdefense += bixo.sp_defense
        pkt.mean_hp += bixo.hp
        pkt.mean_total += bixo.total
        pkt.mean_speed += bixo.speed
    pkt.mean_attack = pkt.mean_attack/float(counter)
    pkt.mean_sattack = pkt.mean_sattack/float(counter)
    pkt.mean_defense = pkt.mean_defense/float(counter)
    pkt.mean_sdefense = pkt.mean_sdefense/float(counter)
    pkt.mean_hp = pkt.mean_hp/float(counter)
    pkt.mean_total = pkt.mean_total/float(counter)
    pkt.mean_speed = pkt.mean_speed/float(counter)
    # standard deviation
    for bixo in poketipos[tipo]:
        pkt.sd_attack += (bixo.attack - pkt.mean_attack) ** 2
        pkt.sd_sattack += (bixo.sp_attack - pkt.mean_sattack) ** 2
        pkt.sd_defense += (bixo.defense - pkt.mean_defense) ** 2
        pkt.sd_sdefense += (bixo.sp_defense - pkt.mean_sdefense) ** 2
        pkt.sd_hp += (bixo.hp - pkt.mean_hp) ** 2
        pkt.sd_speed += (bixo.speed - pkt.mean_speed) ** 2
        pkt.sd_total += (bixo.total - pkt.mean_total) ** 2
    pkt.sd_attack = math.sqrt(pkt.sd_attack / float(counter - 1))
    pkt.sd_sattack = math.sqrt(pkt.sd_sattack / float(counter - 1))
    pkt.sd_defense = math.sqrt(pkt.sd_defense / float(counter - 1))
    pkt.sd_sdefense = math.sqrt(pkt.sd_sdefense / float(counter - 1))
    pkt.sd_hp = math.sqrt(pkt.sd_hp / float(counter - 1))
    pkt.sd_speed = math.sqrt(pkt.sd_speed / float(counter - 1))
    pkt.sd_total = math.sqrt(pkt.sd_total / float(counter - 1))
    poketipos_stats[tipo] = pkt

print("Type stats done")
```

Ok. Now let's plot some of this data and have an idea about what we are dealing with.

First step: distribution ok monsters between the monster types

```python
labels = poketipos.keys()
counter = 0
sizes = []
for tipo in poketipos.keys():
    sizes.append(len(poketipos[tipo]))
    counter += len(poketipos[tipo])
    
plt.rcParams['figure.figsize'] = (20.0, 10.0)
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = len(labels) * (0,)  # dont explode anything
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90, colors=colors)
plt.axis('equal')

print("Total number of monsters: " + str(counter))
```

OK. What about one bar chart for all stats and types?

```python
sp_attack_agg = []
sp_defense_agg = []
attack_agg = []
defense_agg =[]
hp_agg = []
speed_agg =[]
sp_attack_sd = []
sp_defense_sd = []
attack_sd = []
defense_sd =[]
hp_sd = []
speed_sd =[]

for tipo in serializable_labels:
    sp_attack_agg.append(poketipos_stats[tipo].mean_sattack)
    sp_defense_agg.append(poketipos_stats[tipo].mean_sdefense)
    attack_agg.append(poketipos_stats[tipo].mean_attack)
    defense_agg.append(poketipos_stats[tipo].mean_defense)
    hp_agg.append(poketipos_stats[tipo].mean_hp)
    speed_agg.append(poketipos_stats[tipo].mean_speed)
    sp_attack_sd.append('sd: ' + str(poketipos_stats[tipo].sd_sattack))
    sp_defense_sd.append('sd: ' + str(poketipos_stats[tipo].sd_sdefense))
    attack_sd.append('sd: ' + str(poketipos_stats[tipo].sd_attack))
    defense_sd.append('sd: ' + str(poketipos_stats[tipo].sd_defense))
    hp_sd.append('sd: ' + str(poketipos_stats[tipo].sd_hp))
    speed_sd.append('sd: ' + str(poketipos_stats[tipo].sd_speed))

trace1 = go.Bar(
    x=serializable_labels,
    y=sp_attack_agg,
    text=sp_attack_sd,
    name='sp attack'
)
trace2 = go.Bar(
    x=serializable_labels,
    y=sp_defense_agg,
    text=sp_defense_sd,
    name='sp defense'
)
trace3 = go.Bar(
    x=serializable_labels,
    y=attack_agg,
    text=attack_sd,
    name='attack'
)
trace4 = go.Bar(
    x=serializable_labels,
    y=defense_agg,
    text=defense_sd,
    name='defense'
)
trace5 = go.Bar(
    x=serializable_labels,
    y=hp_agg,
    text=hp_sd,
    name='hp'
)
trace6 = go.Bar(
    x=serializable_labels,
    y=speed_agg,
    text=speed_sd,
    name='speed'
)

data = [trace1, trace2, trace3, trace4, trace5, trace6]
layout = go.Layout(
    barmode='stack'
)

fig = go.Figure(data=data, layout=layout)
#py.iplot(fig, filename='stacked-bar')
tls.embed("https://plot.ly/~phboueke/68/")
```

Now just **use the legend to select which stats you want to compare** and hover to see the exact mean and standard deviation values!

```python
# ideally one trace for each type, but we will use text to identify each type
trace1 = dict(type='scatterternary',
              text=serializable_labels,
              a=hp_agg,
              b=attack_agg,
              c=defense_agg,
              mode='markers',
              marker={'symbol': 100,
                      'color': 'red',
                      'size': 10},
)

layout = {
    'title': 'Ternary Scatter Plot: ',
    'ternary':
        {
        'sum':1,
        'aaxis':{'title': 'HP', 'min': 0.01, 'linewidth':2, 'ticks':'outside' },
        'baxis':{'title': 'Attack', 'min': 0.01, 'linewidth':2, 'ticks':'outside' },
        'caxis':{'title': 'Defense', 'min': 0.01, 'linewidth':2, 'ticks':'outside' }
    },
        'showlegend': True,
        'autosize': True,
}

scatter_fig = dict(data=[trace1], layout=layout)
#py.iplot(scatter_fig)
tls.embed("https://plot.ly/~phboueke/78/")
```
